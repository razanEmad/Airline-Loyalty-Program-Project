### Replacing Nulls in Age column with the average

```python
df["AGE"].isnull().any()
```

```python
df.loc[df["AGE"].isna(), "AGE"] = df["AGE"].mean()
```

```python
df["AGE"] = df["AGE"].astype("int")
```

```python
df["AGE"].isnull().any()
```
